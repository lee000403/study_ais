```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time
```

```python
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://emart.ssg.com/disp/category.ssg?dispCtgId=6000214033&page=1")
```

```python
item_list = driver.find_elements(By.XPATH, "//*[@class='mnemitem_goods_tit']")
item_out=[]
for i in range(2,100):
    try:
        for j in range(len(item_list)):
            item_out.append(driver.find_elements(By.XPATH, "//*[@class='mnemitem_goods_tit']")[j].get_attribute("innerText")) 
        driver.find_element(By.XPATH, "//*[@onclick='itemLister.changePage({});return false;']".format(i)).click()
        time.sleep(3)
        
    except:
        break
        
print(item_out)
```

```python
item_list = driver.find_elements(By.XPATH, "//*[@class='mnemitem_goods_tit']")
pic_list = driver.find_elements(By.XPATH, "//*[@class='mnemitem_thmb_link clickable']")
item_out=[]
for i in range(2,3):
    try:
        for j in range(len(item_list)):
            item_out.append(pic_list[j].get_attribute("href"))
            item_out.append(driver.find_elements(By.XPATH, "//*[@class='mnemitem_goods_tit']")[j].get_attribute("innerText")) 
        driver.find_element(By.XPATH, "//*[@onclick='itemLister.changePage({});return false;']".format(i)).click()
        time.sleep(3)
        
    except:
        break
print(item_out)
```
