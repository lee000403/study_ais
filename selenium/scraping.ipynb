```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time
```

```python
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://a73.koreanz.xyz/bbs/board.php?bo_table=tvdrama&page=1")
```

```python
len_item = driver.find_elements(By.XPATH, "//*[@class='list-body']/li")
number = driver.find_elements(By.XPATH, '//*[@id="fboardlist"]/div[1]/ul/li[1]/div')
title = driver.find_elements(By.XPATH, "//*[@class='item-subject']")
datetime = driver.find_elements(By.XPATH, "//*[@class='orangered']")
datetime2 = driver.find_elements(By.XPATH, "//*[@class='wr-date hidden-xs']")
views = driver.find_elements(By.XPATH, "//*[@class='wr-hit hidden-xs']")
program_list = []


for page in range(1, 11):
    driver.get("https://a73.koreanz.xyz/bbs/board.php?bo_table=tvdrama&page={}".format(page))
    time.sleep(5)
    for i in range(len(len_item)):
        try:
            program_listun = []
            program_listun.append(driver.find_element(By.XPATH, '//*[@id="fboardlist"]/div[1]/ul/li[{}]/div[1]'.format(i+1)).text)
            program_listun.append(driver.find_element(By.XPATH, '//*[@id="fboardlist"]/div[1]/ul/li[{}]/div[2]/a'.format(i+1)).text)
            program_listun.append(driver.find_element(By.XPATH, '//*[@id="fboardlist"]/div[1]/ul/li[{}]/div[4]'.format(i+1)).text)
            program_listun.append(driver.find_element(By.XPATH, '//*[@id="fboardlist"]/div[1]/ul/li[{}]/div[5]'.format(i+1)).text)

            driver.find_element(By.XPATH, '//*[@id="fboardlist"]/div[1]/ul/li[{}]/div[2]/a'.format(i+1)).click()
            time.sleep(3)
            link_drama = driver.find_elements(By.XPATH, "//*[@target='_blank']")
            for j in range(len(link_drama)):
                program_listun.append(link_drama[j].get_attribute("href"))
            driver.back()
            time.sleep(3)
            program_list.append(program_listun)   
        except:
            program_listun = []
            program_listun.append(driver.find_element(By.XPATH, '//*[@id="fboardlist"]/div[1]/ul/li[{}]/div[1]'.format(i+1)).text)
            program_listun.append(driver.find_element(By.XPATH, '//*[@id="fboardlist"]/div[1]/ul/li[{}]/div[2]/a'.format(i+1)).text)
            program_listun.append(driver.find_element(By.XPATH, '//*[@id="fboardlist"]/div[1]/ul/li[{}]/div[4]'.format(i+1)).text)
            program_listun.append(driver.find_element(By.XPATH, '//*[@id="fboardlist"]/div[1]/ul/li[{}]/div[5]'.format(i+1)).text)
            program_list.append(program_listun)

            driver.find_element(By.XPATH, '//*[@id="fboardlist"]/div[1]/ul/li[{}]/div[2]/a'.format(i+1)).click()
            time.sleep(3)
            link_drama = driver.find_elements(By.XPATH, "//*[@target='_blank']")
            for j in range(len(link_drama)):
                program_listun.append(link_drama[j].get_attribute("href"))
            driver.back()
            time.sleep(3)
            program_list.append(program_listun)

print(program_list)
```
